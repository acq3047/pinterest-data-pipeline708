```python
# Example transformations for df_pin DataFrame
df_pin_cleaned = df_pin_pinterest.select(
    [when(col(c).isin('', ' ', 'N/A', 'None', None), None).otherwise(col(c)).alias(c) for c in df_pin.columns]
)

# Show the cleaned DataFrame
display(df_pin_cleaned)
```

```python
# Replace empty values in follower_count with 0.
df_pin_clean = df_pin_pinterest.na.fill({'follower_count':0})
display(df_pin_clean)
```

```python
#Perform the necessary transformations on the follower_count to ensure every entry is a number. Make sure the data type of this column is an int.
from pyspark.sql.types import IntegerType

def convert_follower_count(follower_count):
    follower_count = str(follower_count).lower().strip()
    #follower_count = follower_count.lower()
    if 'k' in follower_count:
        return int(float(follower_count.replace('k', '')) * 1000)
    elif 'm' in follower_count:
        return int(float(follower_count.replace('m', '')) * 1000000)
    else:
        try:
            return int(follower_count)
        except ValueError:
            return None


# Register the function as a UDF
convert_follower_count_udf = udf(convert_follower_count, IntegerType())

# Apply the conversion to follower_count and ensure it's an integer
df_pin_clean = df_pin_clean.withColumn(
    "follower_count", convert_follower_count_udf(col("follower_count"))
)

# Show the cleaned DataFrame with the transformed follower_count
display(df_pin_clean)
```

```python
# Ensure that each column containing numeric data has a numeric data type
df_pin_clean.printSchema()
```

```python
# Make the null values in dowloaded = 0
df_pin_clean = df_pin_clean.na.fill({'downloaded':0})
display(df_pin_clean)
```

```python
# Convert the downloaded column to an integer
df_pin_clean = df_pin_clean.withColumn('downloaded', col('downloaded').cast('int'))
```

```python
df_pin_clean.printSchema()
```

```python
# Make the null values in the index column = 0
df_pin_clean = df_pin_clean.na.fill({'index': 0})
display(df_pin_clean)
```

```python
# Convert the index column to an integer
df_pin_clean = df_pin_clean.withColumn('index', col('index').cast('int'))
```

```python
df_pin_clean.printSchema()
```

```python
# Clean the data in the save_location column to include only the save location path
df_pin_clean = df_pin_clean.withColumn('save_location', regexp_replace('save_location', 'Local save in',''))
display(df_pin_clean)
```

```python
# Rename the index column to ind.
df_pin_clean = df_pin_clean.withColumnRenamed('index', 'ind')
df_pin_clean.printSchema()
```

```python
"""
Reorder the DataFrame columns to have the following column order:
    -ind
    -unique_id
    -title
    -description
    -follower_count
    -poster_name
    -tag_list
    -is_image_or_video
    -image_src
    -save_location
    -category
"""
# Reorder the DataFrame columns
desired_columns = [
    "ind", "unique_id", "title", "description", "follower_count", "poster_name",
    "tag_list", "is_image_or_video", "image_src", "save_location", "category"
]

df_pin_clean_reordered = df_pin_clean.select(*desired_columns)
display(df_pin_clean_reordered)
```

## Task 4: Find the most popular category in each country

Find the most popular Pinterest category people post to based on their country.

Your query should return a DataFrame that contains the following columns:

  - country
  - category
  - category_count, a new column containing the desired query output

```python
%sql
--- Find the the most popular category in each country in the DataFrame
SELECT p.category,
    g.country,
    count(p.category) AS category_count --- Count the number of people post to each country
From `df_pin_pinterest` p
    JOIN `df_geo_pinterest` g ON p.ind = g.ind
GROUP by 
    p.category,
    g.country
ORDER BY category_count DESC;

```

## Task 5: Find the most popular category each year

Find how many posts each category had between 2018 and 2022.

Your query should return a DataFrame that contains the following columns:

  - post_year, a new column that contains only the year from the timestamp column
  - category
  - category_count, a new column containing the desired query output

```python
%sql
SELECT
  EXTRACT(YEAR FROM g.timestamp) AS post_year,
  p.category,
  COUNT(p.category) AS category_count
FROM `df_pin_pinterest` p
  JOIN `df_geo_pinterest` g ON p.ind = g.ind
WHERE 
    EXTRACT(YEAR FROM g.timestamp) BETWEEN 2018 AND 2022
GROUP BY 
  EXTRACT(YEAR FROM g.timestamp),
  p.category
ORDER BY post_year DESC

```

## Task 6: Find the user with most followers in each country

**Step 1**: For each country find the user with the most followers.

Your query should return a DataFrame that contains the following columns:

  - country
  - poster_name
  - follower_count

**Step 2**: Based on the above query, find the country with the user with most followers.


Your query should return a DataFrame that contains the following columns:

  - country
  - follower_count

```python
%sql
WITH max_country_followers AS(
  SELECT g.country,
    p.poster_name,
    MAX(p.follower_count) AS follower_count
  FROM `df_pin_pinterest` p
    JOIN `df_geo_pinterest` g  ON  p.ind = g.ind
  GROUP BY g.country,
    p.poster_name
)

SELECT country,
  MAX(follower_count) AS max_followers
FROM max_country_followers
GROUP BY country
ORDER BY max_followers DESC
```

## Task 7: Find the most popular category for different age groups

What is the most popular category people post to based on the following age groups:

  - 18-24
  - 25-35
  - 36-50
  - +50
Your query should return a DataFrame that contains the following columns:

  - age_group, a new column based on the original age column
  - category
  - category_count, a new column containing the desired query output

```python
%sql
WITH pop_cat_age AS (
  SELECT
    CASE
      WHEN u.age BETWEEN 18 AND 24 THEN '18-24'
      WHEN u.age BETWEEN 25 AND 35 THEN '25-35'
      WHEN u.age BETWEEN 36 AND 50 THEN '36-50'
      WHEN u.age > 50 THEN '50+'
    END AS age_group,
    p.category,
    COUNT(p.category) AS category_count
  FROM `df_pin_pinterest` p 
    JOIN `df_user_pinterest` u ON p.ind = u.ind
  GROUP BY age_group,
    p.category
)

SELECT *
FROM pop_cat_age
ORDER BY age_group, category_count DESC
```

## Task 8: Find the median follower count for different age groups

What is the median follower count for users in the following age groups:

  - 18-24
  - 25-35
  - 36-50
  - +50
Your query should return a DataFrame that contains the following columns:

  - **age_group**, a new column based on the original age column
  - **median_follower_count**, a new column containing the desired query output

```python
%sql

WITH med_foll_age_count AS (
  SELECT
    CASE
      WHEN u.age BETWEEN 18 AND 24 THEN '18-24'
      WHEN u.age BETWEEN 25 AND 35 THEN '25-35'
      WHEN u.age BETWEEN 36 AND 50 THEN '36-50'
      WHEN u.age > 50 THEN '50+'
    END AS age_group,
    percentile_approx(p.follower_count, 0.5) AS median_follower_count
  FROM `df_pin_pinterest` p 
    JOIN `df_user_pinterest` u ON p.ind = u.ind
  GROUP BY age_group
)

SELECT *
FROM med_foll_age_count
ORDER BY median_follower_count DESC
```

## Task 9: Find how many users have joined each year?

Find how many users have joined between 2015 and 2020.

Your query should return a DataFrame that contains the following columns:

  - **post_year**, a new column that contains only the year from the timestamp column
  - **number_users_joined**, a new column containing the desired query output

```python
%sql
SELECT
  extract(YEAR FROM g.timestamp) AS post_year,
  COUNT(u.date_joined) AS number_users_joined
FROM `df_geo_pinterest` g 
  JOIN `df_user_pinterest` u ON g.ind = u.ind
WHERE EXTRACT(YEAR FROM g.timestamp) BETWEEN 2018 AND 2022
GROUP BY post_year
ORDER BY number_users_joined DESC
```

## Task 11: Find the median follower count of users based on their joining year and age group

Find the median follower count of users that have joined between 2015 and 2020, based on which age group they are part of.

Your query should return a DataFrame that contains the following columns:

  - **age_group**, a new column based on the original age column
  - **post_year**, a new column that contains only the year from the timestamp column
  - **median_follower_count**, a new column containing the desired query output

```python
%sql
WITH med_foll_age_count AS (
  SELECT
    CASE
      WHEN u.age BETWEEN 18 AND 24 THEN '18-24'
      WHEN u.age BETWEEN 25 AND 35 THEN '25-35'
      WHEN u.age BETWEEN 36 AND 50 THEN '36-50'
      WHEN u.age > 50 THEN '50+'
    END AS age_group,
    EXTRACT(YEAR FROM g.timestamp) AS post_year,
    PERCENTILE_APPROX(p.follower_count, 0.5) AS median_follower_count
    FROM `df_pin_pinterest` p
      JOIN `df_geo_pinterest` g ON p.ind = g.ind
      JOIN `df_user_pinterest` u ON p.ind = u.ind
    WHERE EXTRACT(YEAR FROM u.date_joined) BETWEEN 2015 AND 2020
    GROUP BY age_group, post_year
)

SELECT *
FROM med_foll_age_count
ORDER BY median_follower_count DESC
```
